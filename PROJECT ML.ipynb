```python
import pandas as pd 
import re
from bs4 import BeautifulSoup as BS
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
driver = []

#storing path of chrome driver in "path" variable
path = "C:\webdrivers.exe"
driver.append(webdriver.Chrome(path,options=Options()))
```

```python
inputVar = input("Enter the product for scrapping: ")

#dividing url into 2 parts and adding the item for scrapping

url1 = "https://www.daraz.pk/catalog/?q="
url2 = "&_keyori=ss&from=input&spm=a2a0e.home.search.go.35e34937dlXJOS"
product = re.sub("[ ]", "+", inputVar)
url = url1 + product + url2

#taking user input for number of items and converting it into integer

print("Please do not enter more than 40 because 1 page have 40 products")
NoOfItems = input("How many Items you want to scrap :")
NoOfItems = int(NoOfItems)

#opening url
driver[0].get(url)
pg = driver[0].page_source
soup = BS(pg, "lxml")
```

```python
#finding the total sellers

totalSellers = soup.find("div", {"class": "c1DXz4"})
temp = totalSellers.find('span')
total = re.sub("[^0-9]", '', temp.text)
total = int(total)
#print("Total Sellers: ",total)
```

```python

# finding average price of products
prdPrice = []
average = 0
PrdPrice = soup.find_all("div", {"class": "c3gUW0"})

#getting prices of products
for i in range(len(PrdPrice)):
    temp = PrdPrice[i].find('span')
    prdPrice.append(temp.text)
    
prdPrice = prdPrice[0:NoOfItems]

#calculating average of price
for i in range(len(prdPrice)):
    prdPrice[i] = re.sub("[^0-9]","",prdPrice[i])
    prdPrice[i] = int(prdPrice[i])
    average = average + prdPrice[i]

average = average/NoOfItems

print("\nProduct Prices: ", prdPrice)



#finding average reviews of products
averageRev = 0
rev = soup.find_all("div",{"class":"c15YQ9"})
Rev = []

for i in range(len(rev)):
    temp = rev[i].find('span')
    Rev.append(temp.text)
Rev = Rev[0:NoOfItems]

for i in range(len(Rev)):
    Rev[i] = re.sub("[^0-9]", '', Rev[i])    
    Rev[i]=int(Rev[i])   
    averageRev = averageRev + Rev[i]
    
averageRev = averageRev/NoOfItems
print("\nReviews: ",Rev)


#finding average ratings of products

avgStars = 0
Stars = []
star = soup.find_all("div", {"class": "c2JB4x c6Ntq9"})

for i in star:
    st0 = len(i.find_all("i", {"class": "c3dn4k c1dtTC"}))
    st1 = len(i.find_all("i", {"class": "c3dn4k c1Zozd"}))
    st2 = len(i.find_all("i", {"class": "c3dn4k cbDGcO"}))
    st3 = len(i.find_all("i", {"class": "c3dn4k c17YMy"}))
    st4 = len(i.find_all("i", {"class": "c3dn4k c3fsPU"}))
    st5 = len(i.find_all("i", {"class": "c3dn4k c1e2gb"}))
    st6 = len(i.find_all("i", {"class": "c3dn4k c3An30"}))
    st7 = len(i.find_all("i", {"class": "c3dn4k c3DcGB"}))
    st8 = len(i.find_all("i", {"class": "c3dn4k c1wCjy"})) 
    st9 = len(i.find_all("i", {"class": "c3dn4k cF1vkb"}))
    st10= len(i.find_all("i", {"class": "c3dn4k c3EEAg"}))
  
    Stars.append((st0 + 0.00) + (st1 * 0.10) +
                  (st2 * 0.20) + (st3 * 0.93) +
                  (st4 * 0.40) + (st5 * 0.50) +
                  (st6 * 0.60) + (st7 * 0.70) +
                  (st8 * 0.80) + (st9 * 0.90) + st10)

Stars = Stars[0:NoOfItems]

for i in range(len(Stars)):
    Stars[i] = float(Stars[i])
    avgStars =  avgStars + Stars[i]
    Stars[i] = "{:.2f}".format(Stars[i])
avgStars = avgStars/NoOfItems
print("\nRatings: ",Stars)
```

```python
#For displaying the table finding the following

#finding name of products
Name = []
ProductName = soup.find_all("div", {"class": "c16H9d"})
for i in range (len(ProductName)):
    #as class for product name have "a"
    temp = ProductName[i].find('a')
    Name.append(temp.get('title'))
Name = Name[0:NoOfItems]
print("Product Names: ",Name)


#finding seller country of products
Country = []
SellerCountry = soup.find_all("span", {"class": "c2i43-"})

for i in range(len(SellerCountry)):
    Country.append(SellerCountry[i].text)
Country = Country[0:NoOfItems]
print("\nSeller Country: ",Country)    
   
#finding if the product is in Daraz Mall or not
DrzMall = []
DarazMall = soup.find_all("div", {"class": "c3vCyH"})

for i in range(len(DarazMall)):
    #as class for daraz mall logo have "i"
    temp = DarazMall[i].find('i') 
    if temp is None:
        DrzMall.append("No")
    else:
        DrzMall.append("Yes")
DrzMall = DrzMall[0:NoOfItems] 
print("\nDaraz Mall: ",DrzMall)
        
    
#finding the discount prices of products

discount = soup.find_all("span", {"class": "c1-B2V"})
DisPrice = []
disAvg = 0

for i in range (len(discount)):
    temp = discount[i].find('del')
    DisPrice.append(temp.text)
    
discount = discount[0:NoOfItems]

for i in range(len(DisPrice)):
    DisPrice[i] = re.sub("[^0-9]", '',DisPrice[i])
    DisPrice[i] = int(DisPrice[i])
    disAvg = disAvg + DisPrice[i]
    
disAvg = disAvg/NoOfItems
print("\nDiscounted Prices: ",DisPrice)    
```

```python
print("Total Sellers: ",total)
print("Avg Price: ", average)
print("Avg Reviews: ","{:.2f}".format(averageRev))
print("Avg Rating: ","{:.2f}".format(avgStars))
```
